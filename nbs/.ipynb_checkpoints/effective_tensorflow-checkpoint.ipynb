### Symbolism in Tensorflow

```python
import numpy as np

x = np.random.normal(size=[10, 10])
y = np.random.normal(size=[10, 10])
z = np.dot(x, y)

print(z)
```

```python
import tensorflow as tf
x = tf.random_normal([10,10])
y = tf.random_normal([10,10])
z = tf.matmul(x, y)

sess = tf.Session()
z_val = sess.run(z)
print(z_val)
```

#### Unlike np the tensorflow only gives us a handle to a node in the graph that represents the result
